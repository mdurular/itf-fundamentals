```python
fruits = ['Banana', 'Orange', 'Apple', 'Strawberry', 'Cherry']

with open("fruits.txt", 'w', encoding="utf-8") as file:
    for i in fruits:
        file.write(i + '\n')    
with open("fruits.txt", 'r', encoding="utf-8") as file:
    print(file.read())
```

```python
counter=0
with open("istiklal.txt", 'w', encoding="utf-8") as file :
        for i in lines:
            counter+=1
            if counter%4==0:
                file.write(i + "\n")
            else:
                file.write(i)
                
with open("istiklal.txt", 'r', encoding="utf-8") as f :
    new=f.read()
```

```python
with open("istiklal.txt", 'a', encoding="utf-8") as file :
        file.writelines(lines)
with open("istiklal.txt", 'r', encoding="utf-8") as f :
    print(f.read())
```

```python
with open("istiklal.txt", 'r', encoding="utf-8") as file :
    new_lines=file.readlines()
```

```python
with open("istiklal.txt", 'a', encoding="utf-8") as f :
        [i.replace("\n", "") if i == "\n" else f.write(i) for i in new_lines]
with open("istiklal.txt", 'r', encoding="utf-8") as f :
    print(f.read())
```
